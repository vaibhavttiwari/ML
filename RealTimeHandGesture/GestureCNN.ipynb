```python
import numpy as np
import os
from PIL import Image
import matplotlib.pyplot as plt
import keras
from sklearn.model_selection import train_test_split
```

```python
X_data = []
Y_data = []

for i in os.listdir("dataset/"):
    for j in os.listdir("dataset/"+i+"/"):
        print("Reading : dataset/"+i+"/"+j)
        img = Image.open("dataset/"+i+"/"+j).convert('L')
        arr = np.array(img)
        
        X_data.append(arr)
        Y_data.append(int(i))
```

```python
X_data = np.array(X_data, dtype='float32')
Y_data = np.array(Y_data, dtype='int32')
```

```python
X_data /= 255
X_data = X_data.reshape((1505, 200, 200, 1))
Y_data = Y_data.reshape((1505,1))
print(X_data.shape)
print(Y_data.shape)
```

```python
Y_data = keras.utils.to_categorical(Y_data)

X_train, X_temp, Y_train, Y_temp = train_test_split(X_data, Y_data, test_size = 0.2)
X_test, X_val, Y_test, Y_val = train_test_split(X_temp, Y_temp, test_size = 0.5)
```

```python
Y_data.shape
```

```python
from keras import models
from keras import layers
```

```python
model = models.Sequential()

model.add(layers.Conv2D(32, (5, 5), strides=(2, 2), activation='relu', input_shape=(200, 200, 1)))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Flatten())

model.add(layers.Dense(128, activation='relu'))

model.add(layers.Dense(64, activation='relu'))

model.add(layers.Dense(5, activation='softmax'))
```

```python
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, batch_size=64, epochs=10, verbose=1, validation_data=(X_val, Y_val))
```

```python
model_json = model.to_json()

with open("model.json", "w") as json_file:
    json_file.write(model_json)
    
model.save_weights("model.h5")
print("Saved")
```
